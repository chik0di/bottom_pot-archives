```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.scrapethissite.com/pages/simple/"

response = requests.get(url)

soup = BeautifulSoup(response.text, 'html')
```

```python
countries = soup.find_all('div', class_= 'col-md-4 country')

atlas = []

for country in countries:
    name = country.find('h3').text.strip()
    capitals = country.find('span', class_= 'country-capital').text
    population = country.find('span', class_= 'country-population')
    geo = int(population.text)
    landmass = country.find('span', class_= 'country-area')
    area = float(landmass.text)
    atlas.append([name, capitals, geo, area])

atlas
```

```python
df = pd.DataFrame(atlas, columns=['Country', 'Capital', 'Population', 'Area'])
df
```

```python
df.to_csv('world_countries.csv')
```
